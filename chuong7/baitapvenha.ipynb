```python
import tkinter as tk

def check_fields():
    if entry_name.get() and entry_id.get() and entry_pass.get():
        btn_submit.pack(pady=10)

root = tk.Tk()
root.title("Thông tin sinh viên")

# Nhãn + Entry
tk.Label(root, text="Tên sinh viên").pack()
entry_name = tk.Entry(root)
entry_name.pack()
entry_name.bind("<KeyRelease>", lambda e: check_fields())

tk.Label(root, text="ID sinh viên").pack()
entry_id = tk.Entry(root)
entry_id.pack()
entry_id.bind("<KeyRelease>", lambda e: check_fields())

tk.Label(root, text="Mật khẩu").pack()
entry_pass = tk.Entry(root, show="*")
entry_pass.pack()
entry_pass.bind("<KeyRelease>", lambda e: check_fields())

# Nút gửi (ẩn ban đầu)
btn_submit = tk.Button(root, text="Gửi")

root.mainloop()
```

```python
import tkinter as tk

def reverse_string():
    text = entry.get()
    result = ""
    for i in range(len(text) - 1, -1, -1):
        result += text[i]
    label_result.config(text=result)

root = tk.Tk()
root.title("Reverse Word")

tk.Label(root, text="Enter a word:").pack(pady=5)

entry = tk.Entry(root, width=30)
entry.pack(pady=5)

tk.Button(root, text="Validate", command=reverse_string).pack(pady=10)

label_result = tk.Label(root, text="", font=("Arial", 12, "bold"))
label_result.pack(pady=5)

root.mainloop()
```

```python
import tkinter as tk

def calculate_sum():
    n = int(entry.get())
    total = 0
    for i in range(1, n + 1):
        total += i
    label_result.config(text=f"Tổng = {total}")

root = tk.Tk()
root.title("Tính tổng 1 + 2 + ... + N")

tk.Label(root, text="Nhập số N:").pack(pady=5)

entry = tk.Entry(root)
entry.pack(pady=5)

tk.Button(root, text="Calculate", command=calculate_sum).pack(pady=10)

label_result = tk.Label(root, text="")
label_result.pack(pady=5)

root.mainloop()
```

```python
import tkinter as tk

def find_divisors():
    n = int(entry.get())
    result = ""
    for i in range(1, n + 1):
        if n % i == 0:
            result += str(i) + " "
    label_result.config(text="The divisors of N : " + result)

root = tk.Tk()
root.title("Divisors of N")

tk.Label(root, text="Enter the value of N").pack(pady=5)

entry = tk.Entry(root)
entry.pack(pady=5)

tk.Button(root, text="Validate", command=find_divisors).pack(pady=10)

label_result = tk.Label(root, text="")
label_result.pack(pady=5)

root.mainloop()
```

```python
import tkinter as tk

def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

def lcm(a, b):
    return a * b // gcd(a, b)

def calculate():
    m = int(entry_m.get())
    n = int(entry_n.get())
    choice = var.get()

    if choice == "GCD":
        result = gcd(m, n)
        label_result.config(text=f"GCD(m, n) = {result}")
    else:
        result = lcm(m, n)
        label_result.config(text=f"LCM(m, n) = {result}")

root = tk.Tk()
root.title("GCD & LCM")

tk.Label(root, text="Enter value of m:").pack()
entry_m = tk.Entry(root)
entry_m.pack()

tk.Label(root, text="Enter value of n:").pack()
entry_n = tk.Entry(root)
entry_n.pack()

var = tk.StringVar(value="GCD")
tk.OptionMenu(root, var, "GCD", "LCM").pack(pady=5)

tk.Button(root, text="Validate", command=calculate).pack(pady=10)

label_result = tk.Label(root, text="")
label_result.pack()

root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

def show_message():
    name = entry_name.get()
    age = int(entry_age.get())

    if age >= 18:
        msg = f"Xin chào {name}\nBạn đã trưởng thành!"
    else:
        msg = f"Xin chào {name}\nBạn còn nhỏ tuổi!"

    messagebox.showinfo("Thông báo", msg)

root = tk.Tk()
root.title("Thông tin người dùng")

tk.Label(root, text="Nhập tên:").pack()
entry_name = tk.Entry(root)
entry_name.pack()

tk.Label(root, text="Nhập tuổi:").pack()
entry_age = tk.Entry(root)
entry_age.pack()

tk.Button(root, text="OK", command=show_message).pack(pady=10)

root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

can = ["Giáp","Ất","Bính","Đinh","Mậu","Kỷ","Canh","Tân","Nhâm","Quý"]
chi = ["Tý","Sửu","Dần","Mão","Thìn","Tỵ","Ngọ","Mùi","Thân","Dậu","Tuất","Hợi"]

def convert_year():
    year = int(entry.get())
    can_chi = can[year % 10] + " " + chi[year % 12]
    messagebox.showinfo("Kết quả", f"Năm âm lịch: {can_chi}")

root = tk.Tk()
root.title("Đổi năm Âm lịch")

tk.Label(root, text="Nhập năm Dương lịch:").pack(pady=5)
entry = tk.Entry(root)
entry.pack(pady=5)

tk.Button(root, text="Chuyển đổi", command=convert_year).pack(pady=10)

root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

def tinh_bmi():
    try:
        can_nang = float(entry_weight.get())
        chieu_cao = float(entry_height.get())
        bmi = can_nang / (chieu_cao ** 2)
        label_bmi.config(text=f"BMI: {bmi:.2f}")

        if bmi < 18.5:
            kq = "Gầy"
        elif bmi < 25:
            kq = "Bình thường"
        elif bmi < 30:
            kq = "Thừa cân"
        else:
            kq = "Béo phì"

        label_ketluan.config(text=f"Kết luận: {kq}")
    except:
        messagebox.showerror("Lỗi", "Vui lòng nhập số hợp lệ")

root = tk.Tk()
root.title("BMI Calculator")
root.geometry("300x200")

tk.Label(root, text="Cân nặng (kg):").pack()
entry_weight = tk.Entry(root)
entry_weight.pack()

tk.Label(root, text="Chiều cao (m):").pack()
entry_height = tk.Entry(root)
entry_height.pack()

tk.Button(root, text="Tính BMI", command=tinh_bmi).pack(pady=5)

label_bmi = tk.Label(root, text="BMI:")
label_bmi.pack()

label_ketluan = tk.Label(root, text="Kết luận:")
label_ketluan.pack()

root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

def hien_thong_bao(chuong, bai):
    messagebox.showinfo(
        "Lập trình hướng đối tượng",
        f"Viết phương trình giải phương trình bậc 2.\n(Bài {chuong}.{bai})"
    )

root = tk.Tk()
root.title("Hệ thống bài tập nâng cao Python")
root.geometry("400x300")

menubar = tk.Menu(root)
root.config(menu=menubar)

for chuong in range(1, 8):
    menu_chuong = tk.Menu(menubar, tearoff=0)
    for bai in range(1, 10):
        menu_chuong.add_command(
            label=f"{chuong}.{bai}",
            command=lambda c=chuong, b=bai: hien_thong_bao(c, b)
        )
    menubar.add_cascade(label=f"Chương {chuong}", menu=menu_chuong)

menubar.add_command(label="Exit", command=root.quit)

root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

# Nhập loại xe
print("Chọn loại xe:")
print("1. Xe 4 chỗ")
print("2. Xe 7 chỗ")
loai_xe = int(input("Nhập lựa chọn (1 hoặc 2): "))

# Nhập quãng đường và thời gian chờ
km = float(input("Nhập số km đã đi: "))
cho = int(input("Nhập số phút chờ: "))

tien = 0

# Giá mở cửa
tien += 12000
km_con_lai = km - 0.8

if km_con_lai < 0:
    km_con_lai = 0

# Tính tiền theo loại xe
if loai_xe == 1:  # Xe 4 chỗ
    if km_con_lai <= 30:
        tien += km_con_lai * 15300
    else:
        tien += 30 * 15300
        tien += (km_con_lai - 30) * 12100

elif loai_xe == 2:  # Xe 7 chỗ
    if km_con_lai <= 30:
        tien += km_con_lai * 16100
    else:
        tien += 30 * 16100
        tien += (km_con_lai - 30) * 13800
else:
    print("Loại xe không hợp lệ!")
    exit()

# Tính tiền chờ
if cho > 5:
    tien += (cho - 5) * 750

# In kết quả
print("Tổng tiền taxi phải trả là:", int(tien), "đồng")
```
