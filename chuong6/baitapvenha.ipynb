```python
import threading
import time

def download(file):
    print(f"Downloading {file}...")
    time.sleep(2)
    print(f"Finished {file}")

files = ["file1.pdf", "file2.mp4", "file3.zip"]

threads = []
for f in files:
    t = threading.Thread(target=download, args=(f,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()
```

```python
import threading

def even():
    for i in range(30, 51):
        if i % 2 == 0:
            print("Even:", i)

def odd():
    for i in range(30, 51):
        if i % 2 != 0:
            print("Odd:", i)

t1 = threading.Thread(target=even)
t2 = threading.Thread(target=odd)

t1.start()
t2.start()

t1.join()
t2.join()
```

```python
import threading
import math

def factorial(n):
    print(f"Factorial {n} =", math.factorial(n))

numbers = [5, 6, 7, 8]

threads = []
for n in numbers:
    t = threading.Thread(target=factorial, args=(n,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()
```

```python
import threading
import time

def print_time(thread_name, delay, repeat):
    print(f"Starting {thread_name}")
    for _ in range(repeat):
        time.sleep(delay)
        print(f"{thread_name}: {time.ctime()}")
    print(f"Exiting {thread_name}")

def main():
    print("Starting Main Thread")

    # Tạo các luồng
    google = threading.Thread(
        target=print_time, args=("Google", 1, 5)
    )
    yahoo = threading.Thread(
        target=print_time, args=("Yahoo", 2, 4)
    )
    facebook = threading.Thread(
        target=print_time, args=("Facebook", 3, 3)
    )

    # Khởi động luồng
    google.start()
    yahoo.start()
    facebook.start()

    # Chờ các luồng chạy xong
    google.join()
    yahoo.join()
    facebook.join()

    print("Exiting Main Thread")

if __name__ == "__main__":
    main()
```

```python
import threading
import random

# Tạo list ngẫu nhiên
lst = [random.randint(0, 10) for _ in range(40)]
print("List:", lst)

result = []
lock = threading.Lock()

def partial_sum(sublist, idx):
    s = sum(sublist)
    with lock:
        result.append(s)
    print(f"Thread {idx} sum:", s)

threads = []
chunk_size = len(lst) // 4

for i in range(4):
    start = i * chunk_size
    end = None if i == 3 else (i + 1) * chunk_size
    t = threading.Thread(
        target=partial_sum,
        args=(lst[start:end], i + 1)
    )
    threads.append(t)
    t.start()

for t in threads:
    t.join()

print("Total sum:", sum(result))
```

```python
import threading
import time

def print_time(name, delay, repeat):
    print(f"Starting {name}")
    for _ in range(repeat):
        time.sleep(delay)
        print(f"{name}: {time.ctime()}")
    print(f"Exiting {name}")

print("Starting Main Thread")

google = threading.Thread(
    target=print_time, args=("Google", 1, 4)
)
yahoo = threading.Thread(
    target=print_time, args=("Yahoo", 2, 3)
)
facebook = threading.Thread(
    target=print_time, args=("Facebook", 3, 3)
)

google.start()
yahoo.start()
facebook.start()

google.join()
yahoo.join()
facebook.join()

print("Exiting Main Thread")
```

```python
import threading
import random

lst = [random.randint(0, 100) for _ in range(40)]
print("List:", lst)

max_values = []
lock = threading.Lock()

def find_max(sublist, idx):
    m = max(sublist)
    with lock:
        max_values.append(m)
    print(f"Thread {idx} max:", m)

threads = []
chunk_size = len(lst) // 4

for i in range(4):
    start = i * chunk_size
    end = None if i == 3 else (i + 1) * chunk_size
    t = threading.Thread(
        target=find_max,
        args=(lst[start:end], i + 1)
    )
    threads.append(t)
    t.start()

for t in threads:
    t.join()

print("Max value in list:", max(max_values))
```
