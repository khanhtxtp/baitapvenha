```python


# 1a) Đọc file stocks1.csv
stocks1 = pd.read_csv('stocks1.csv')
print("--- 5 dòng đầu stocks1 ---")
print(stocks1.head())
print("--- 5 dòng cuối stocks1 ---")
print(stocks1.tail())
print("--- Kiểu dữ liệu stocks1 ---")
print(stocks1.dtypes)
print("--- Thông tin stocks1 ---")
stocks1.info()

# 1b) Đọc file stocks2.csv
stocks2 = pd.read_csv('stocks2.csv')
print("--- 5 dòng đầu stocks2 ---")
print(stocks2.head())
print("--- 5 dòng cuối stocks2 ---")
print(stocks2.tail())
print("--- Kiểu dữ liệu stocks2 ---")
print(stocks2.dtypes)
print("--- Thông tin stocks2 ---")
stocks2.info()

# 1c) Đọc file companies.csv
companies = pd.read_csv('companies.csv')
print("--- 5 dòng đầu companies ---")
print(companies.head())
print("--- Kiểu dữ liệu companies ---")
print(companies.dtypes)
print("--- Thông tin companies ---")
companies.info()

# 2) Xử lý Null
# Thay thế Null cột "low" bằng giá trị max cùng stock
# Thay thế Null cột "high" bằng giá trị min cùng stock
stocks1['low'] = stocks1.groupby('symbol')['low'].transform(lambda x: x.fillna(x.max()))
stocks1['high'] = stocks1.groupby('symbol')['high'].transform(lambda x: x.fillna(x.min()))

stocks2['low'] = stocks2.groupby('symbol')['low'].transform(lambda x: x.fillna(x.max()))
stocks2['high'] = stocks2.groupby('symbol')['high'].transform(lambda x: x.fillna(x.min()))

# 3) Tạo dataframe stocks_companies
stocks = pd.concat([stocks1, stocks2], ignore_index=True)
stocks_companies = stocks.merge(companies, on='symbol', how='left')
print("--- 5 dòng đầu stocks_companies ---")
print(stocks_companies.head())

# 4) Giá trung bình theo từng công ty
avg_price = stocks.groupby('symbol')[['open','high','low','close']].mean()
avg_volume = stocks.groupby('symbol')['volume'].mean()
print("--- Giá trung bình ---")
print(avg_price)
print("--- Volume trung bình ---")
print(avg_volume)

# 5) Giá đóng cửa thấp nhất, cao nhất mỗi công ty
close_stats = stocks.groupby('symbol')['close'].agg(['min','max'])
print("--- Thống kê close mỗi công ty ---")
print(close_stats)

# 6) Tạo cột parsed_time dạng datetime
stocks_companies['parsed_time'] = pd.to_datetime(stocks_companies['date'])
print("--- Kiểu dữ liệu parsed_time ---")
print(stocks_companies['parsed_time'].dtype)
print(stocks_companies.head())

# 7) Tạo cột result: close > open ? 'up' : 'down'
stocks_companies['result'] = stocks_companies.apply(lambda x: 'up' if x['close'] > x['open'] else 'down', axis=1)

print("--- 5 dòng cuối cùng với result ---")
print(stocks_companies.tail())
```
