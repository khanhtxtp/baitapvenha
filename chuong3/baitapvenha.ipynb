```python
import numpy as np

arr_a = np.array([1,2,3,2,3,4,3,4,5,6])
arr_b = np.array([7,2,10,2,7,4,9,4,9,8])

#1
arr_c = np.intersect1d(arr_a, arr_b)
print("arr_c =", arr_c)

#1
arr_d = np.setdiff1d(arr_a, arr_b)
print("arr_d =", arr_d)

#1
arr_e = np.array([2,6,1,9,10,3,27,8,6,25,16])
arr_f = arr_e[(arr_e >= 5) & (arr_e <= 10)]
print("arr_f =", arr_f)
```

```python
#1:
arr = np.arange(10)
print("arr =", arr)
print("Kiểu dữ liệu:", arr.dtype)
print("Kích thước:", arr.shape)

#2
arr_odd = arr[arr % 2 != 0]
arr_even = arr[arr % 2 == 0]
print("arr_odd =", arr_odd)
print("arr_even =", arr_even)

#3
arr_update_1 = np.where(arr % 2 == 0, arr, 100)
print("arr_update_1 =", arr_update_1)
```
