```python
from xml.dom import minidom

# ƒê·ªçc file XML
with open("sample.xml", "r", encoding="utf-8-sig") as f:
    xml_text = f.read().strip()

doc = minidom.parseString(xml_text)

# L·∫•y danh s√°ch t·∫•t c·∫£ c√°c ph·∫ßn t·ª≠ c√≥ th·∫ª <name>
names = doc.getElementsByTagName("name")

print("Danh s√°ch c√°c ph·∫ßn t·ª≠ <name> trong sample.xml:")
for i, name in enumerate(names, start=1):
    # firstChild.data l·∫•y n·ªôi dung gi·ªØa c·∫∑p th·∫ª <name>...</name>
    print(f"{i}. {name.firstChild.data}")
```

```python
import json
from collections import Counter

# ƒê·ªçc d·ªØ li·ªáu t·ª´ file nhanvien.json
with open("nhanvien.json", "r", encoding="utf-8") as f:
    data = json.load(f)

company = data["company"]
employees = data["employees"]

count = Counter(emp["department"] for emp in employees)
total = len(employees)

print(f"T√™n c√¥ng ty: {company['name']}")
print(f"ƒê·ªãa ch·ªâ: {company['address']}")
print("----- Th·ªëng k√™ nh√¢n vi√™n theo ƒë∆°n v·ªã -----")

for i, (dept, num) in enumerate(count.items(), start=1):
    print(f"{i}/T√™n ƒë∆°n v·ªã: ƒê∆°n v·ªã {dept}")
    print(f"   - S·ªë nh√¢n vi√™n: {num}")
    print(f"   - T·ª∑ l·ªá: {num / total * 100:.2f}%\n")
```

```python
import requests

URL = "https://jsonplaceholder.typicode.com/posts"
r = requests.get(URL, timeout=15)
r.raise_for_status()
data = r.json()

print("T·ªïng s·ªë b√†i post:", len(data))
for i, post in enumerate(data, start=1):
    print(f"\n{i}. userId: {post['userId']} | id: {post['id']}")
    print(f"   title: {post['title']}")
    print(f"   body : {post['body']}")
```

```python
import requests

# URL API (em c√≥ th·ªÉ thay b·∫±ng API kh√°c n·∫øu mu·ªën)
url = "https://fakestoreapi.com/products"

# G·ª≠i y√™u c·∫ßu GET ƒë·∫øn API
response = requests.get(url)

# Ki·ªÉm tra xem c√≥ l·ªói kh√¥ng
if response.status_code == 200:
    books = response.json()  # Chuy·ªÉn d·ªØ li·ªáu JSON th√†nh list
    print(f"T·ªïng s·ªë s√°ch n·ªïi b·∫≠t: {len(books)}\n")
    
    # In t·ª´ng s√°ch ra m√†n h√¨nh
    for i, book in enumerate(books, start=1):
        print(f"{i}. T√™n s√°ch: {book['title']}")
        print(f"   Gi√°: {book['price']} USD")
        print(f"   M√¥ t·∫£: {book['description']}")
        print()
else:
    print("‚ö†Ô∏è Kh√¥ng th·ªÉ k·∫øt n·ªëi t·ªõi API!")
```
