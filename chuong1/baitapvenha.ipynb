```python
class HinhChuNhat:
    def __init__(self, dai=0, rong=0):
        self.dai = dai
        self.rong = rong

    def nhap(self):
        self.dai = float(input("Nhập chiều dài: "))
        self.rong = float(input("Nhập chiều rộng: "))

    def tinh_chu_vi(self):
        return 2 * (self.dai + self.rong)

    def tinh_dien_tich(self):
        return self.dai * self.rong

    def in_thong_tin(self):
        print(f"Chiều dài: {self.dai}")
        print(f"Chiều rộng: {self.rong}")
        print(f"Chu vi: {self.tinh_chu_vi()}")
        print(f"Diện tích: {self.tinh_dien_tich()}")


# --- Chương trình chính ---
if __name__ == "__main__":
    hcn = HinhChuNhat()
    hcn.nhap()
    print("\nThông tin hình chữ nhật:")
    hcn.in_thong_tin()
```

```python
class ThiSinh:
    def __init__(self, ho_ten="", toan=0, ly=0, hoa=0):
        self.ho_ten = ho_ten
        self.toan = toan
        self.ly = ly
        self.hoa = hoa

    def nhap(self):
        self.ho_ten = input("Nhập họ tên thí sinh: ")
        self.toan = float(input("Điểm Toán: "))
        self.ly = float(input("Điểm Lý: "))
        self.hoa = float(input("Điểm Hóa: "))

    def tong_diem(self):
        return self.toan + self.ly + self.hoa

    def in_thong_tin(self):
        print(f"Họ tên: {self.ho_ten}")
        print(f"Toán: {self.toan}, Lý: {self.ly}, Hóa: {self.hoa}")
        print(f"Tổng điểm: {self.tong_diem()}")
        print("-" * 30)


# --- Chương trình chính ---
if __name__ == "__main__":
    ds = []
    n = int(input("Nhập số lượng thí sinh: "))

    for i in range(n):
        print(f"\n-- Nhập thí sinh thứ {i+1} --")
        ts = ThiSinh()
        ts.nhap()
        ds.append(ts)

    print("\nDanh sách thí sinh:")
    for ts in ds:
        ts.in_thong_tin()

    # Lọc trúng tuyển
    diem_chuan = float(input("\nNhập điểm chuẩn: "))
    print(f"\nDanh sách thí sinh trúng tuyển (>= {diem_chuan}):")
    for ts in ds:
        if ts.tong_diem() >= diem_chuan:
            print(f"{ts.ho_ten} - Tổng điểm: {ts.tong_diem()}")
```

```python
class Stack:
    def __init__(self, size):
        self.size = size
        self.stack = [0.0] * size
        self.top = -1

    def isEmpty(self):
        return self.top == -1

    def isFull(self):
        return self.top == self.size - 1

    def push(self, value):
        if not self.isFull():
            self.top += 1
            self.stack[self.top] = value
        else:
            print("Ngăn xếp đã đầy!")

    def pop(self):
        if not self.isEmpty():
            value = self.stack[self.top]
            self.top -= 1
            return value
        else:
            print("Ngăn xếp rỗng!")
            return None
```

```python
class StackCount(Stack):
    def __init__(self, size):
        super().__init__(size)
        self.count = 0

    def push(self, value):
        super().push(value)
        if not self.isFull():
            self.count += 1

    def getCount(self):
        return self.count
```

```python
class Date:
    def __init__(self, ngay=1, thang=1, nam=2000):
        self.ngay = ngay
        self.thang = thang
        self.nam = nam

    def display(self):
        print(f"Ngày {self.ngay}/{self.thang}/{self.nam}")

    def next(self):
        self.ngay += 1
        # Giản lược xử lý ngày tháng cho ngắn gọn
        if self.ngay > 30:
            self.ngay = 1
            self.thang += 1
        if self.thang > 12:
            self.thang = 1
            self.nam += 1
```

```python
class Account:
    def __init__(self, so_tk, ten_tk, so_du=0.0):
        self.so_tk = so_tk
        self.ten_tk = ten_tk
        self.so_du = so_du

    def nap_tien(self, so_tien):
        self.so_du += so_tien
        print(f"Nạp {so_tien} thành công. Số dư hiện tại: {self.so_du}")

    def rut_tien(self, so_tien):
        if so_tien <= self.so_du:
            self.so_du -= so_tien
            print(f"Rút {so_tien} thành công. Số dư còn lại: {self.so_du}")
        else:
            print("Số dư không đủ!")

    def hien_thi(self):
        print(f"Số TK: {self.so_tk}, Tên TK: {self.ten_tk}, Số dư: {self.so_du}")
```

```python
class TamGiac:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

    def hop_le(self):
        return self.a + self.b > self.c and self.a + self.c > self.b and self.b + self.c > self.a

    def chu_vi(self):
        if self.hop_le():
            return self.a + self.b + self.c
        return 0

    def dien_tich(self):
        if self.hop_le():
            p = self.chu_vi() / 2
            from math import sqrt
            return sqrt(p * (p - self.a) * (p - self.b) * (p - self.c))
        return 0

    def in_tam_giac(self):
        if self.hop_le():
            print(f"Tam giác có cạnh ({self.a}, {self.b}, {self.c})")
            print(f"Chu vi: {self.chu_vi()}, Diện tích: {self.dien_tich():.2f}")
        else:
            print("Không phải tam giác hợp lệ!")
```
